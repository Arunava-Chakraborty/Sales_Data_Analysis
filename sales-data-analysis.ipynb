```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/d/aungpyaeap/supermarket-sales/supermarket_sales - Sheet1.csv')

df.head()
```

```python
#df.isnull().sum()

#df.info()



df.describe()
```

# ****Exploratory Data Analysis(EDA)****

```python
#Total Sales Overtime
sales_over_time = df.groupby('Date')['Total'].sum()

plt.figure(figsize=(14 ,6))
sales_over_time.plot()
plt.title('Total sales over time')
plt.xlabel('Date')
plt.ylabel('sales(Total)')
plt.show()
```

```python
branch_sales = df.groupby('Branch')['Total'].sum().sort_values()

plt.figure(figsize = (8,5))
branch_sales.plot(kind='bar', color =['#FF9999','#66B2FF','#99FF99'])
plt.title("Total Sales By Branch")
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()


```

```python
# Sales by Payment Method
payment_sales = df.groupby('Payment')['Total'].sum()

plt.figure(figsize=(8, 5))
payment_sales.plot(kind='bar', color=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Total Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Sales by Product Line
product_sales = df.groupby('Product line')['Total'].sum().sort_values()

plt.figure(figsize=(10, 6))
product_sales.plot(kind='barh', color='skyblue')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

```python
# Average Rating by Product Line
avg_rating_product = df.groupby('Product line')['Rating'].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_rating_product.plot(kind='barh', color='salmon')
plt.title('Average Rating by Product Line')
plt.xlabel('Average Rating')
plt.ylabel('Product Line')
plt.show()
```

```python
# Sales by Gender
gender_sales = df.groupby('Gender')['Total'].sum()

plt.figure(figsize=(8, 5))
gender_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Sales by Gender')
plt.ylabel('')
plt.show()
```

```python
# Sales by Customer Type
customer_sales = df.groupby('Customer type')['Total'].sum()

plt.figure(figsize=(8, 5))
customer_sales.plot(kind='bar', color=['#8fbc8f', '#6495ed'])
plt.title('Total Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Overall summary of insights
print("Insights and Conclusions:")
print("1. The busiest branch with the highest sales is identified by analyzing the 'Branch' column.")
print("2. Sales distribution among different product lines shows which categories are the most popular.")
print("3. Analysis of payment methods reveals the most commonly used payment type.")
print("4. Average rating by product line gives insights into customer satisfaction for each category.")
print("5. Gender distribution shows if there is a demographic that dominates the purchasing activity.")
print("6. Customer type analysis differentiates between normal customers and members, showing members tend to spend more on average.")
```
